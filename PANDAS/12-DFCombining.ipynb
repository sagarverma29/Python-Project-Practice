```python
#DATAFRAME COMBINING
import pandas as pd
import numpy as np
```

```python
data1 = np.full((5,5),1)
index1 = (0,1,2,3,4)
columns1 = (0,1,2,3,4)
```

```python
data2 = np.full((5,5),2)
index2 = (5,6,7,8,9)
columns2 = (0,1,2,3,4)
```

```python
data3 = np.full((5,5),3)
index3 = (0,1,2,3,4)
columns3 = (5,6,7,8,9)
```

```python
df1 = pd.DataFrame(data1, index1, columns1)
df2 = pd.DataFrame(data2, index2, columns2)
df3 = pd.DataFrame(data3, index3, columns3)
```

```python
#pd.concat to combine dataframes
pd.concat([df1,df2], axis=1)
```

```python
pd.concat([df1,df3], axis=1)
```

```python
pd.concat([df2,df3], axis=0)
```
