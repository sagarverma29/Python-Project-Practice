```python
import pandas as pd
countries_df = pd.read_csv(r"C:\Users\SAGAVERM\Desktop\DA dataset\top_10_countries.csv")
```

```python
countries_df.head()
```

```python
countries_df.dtypes
```

```python
countries_df.drop(columns=['% of world', 'Date'], inplace=True)
```

```python
countries_df.rename(columns={'Country / Dependency':'Country'}, inplace=True)
```

```python
countries_df[['Country','Region']] = countries_df[['Country', 'Region']].astype('string')
```

```python
countries_df['Population'] = countries_df['Population'].str.replace(',','')
countries_df['Population'] = countries_df['Population'].astype(int)
```

```python
countries_df.dtypes
```

```python
subregion = ['Eastern Asia', 'Southern Asia','Northern America','Southeast Asia','Southern Asia',
             'Southern America','Western Africa', 'Southern Asia','Eastern Europe','Central America']
```

```python
countries_df['SubRegion'] = subregion
```

```python
#reindexing columns
countries_df=countries_df.reindex(columns=['Rank', 'Country','Region','SubRegion','Population'])
```

```python
countries_df
```

```python
#AGGREGATE FUNCTIONS

print(f"sum: {countries_df['Population'].sum()}")
print(f"max: {countries_df['Population'].max()}")
print(f"min: {countries_df['Population'].min()}")
print(f"mean: {countries_df['Population'].mean()}")
print(f"standard deviation: {countries_df['Population'].std()}")
print(f"count: {countries_df['Population'].count()}")
```

```python
#describe function returns all the aggregate value for a column
countries_df['Population'].describe()
```

```python
#if describe is applied to the whole dataframe, it will return aggregate values only for numnerical columns
countries_df.describe()
```

```python
#while other aggregate functions cannnot be applied to strings, max and min can be applied
print(countries_df['SubRegion'].max())
print(countries_df['SubRegion'].min())
```

```python
countries_df['Region'].unique()
```

```python
countries_df['Region'].nunique()
```

```python
#GROUP BY
countries_df.groupby(by=['Region']) # return an object
```

```python
countries_df.groupby(by=['Region']).sum() #can use different aggregate functions like min, max, mean also
```

```python
countries_df.groupby(by=['Region']).sum(numeric_only=True) #returns only numeric columns
```

```python
#can select the rows that will be returned
countries_df[['Region','Population']].groupby(by=['Region']).max()
```

```python
#can perform groupby on multiple columns
countries_df[['Region','SubRegion','Population']].groupby(by=['Region','SubRegion']).sum()
```

```python
#agg() function to return multiple aggregate function 
countries_df[['Region','SubRegion','Population']].groupby(by=['Region','SubRegion']).agg(['sum','max'])
```

```python
#pivot_table function()
pd.pivot_table(data = countries_df, index = ['Region','Country'], columns='SubRegion', values='Population', aggfunc='sum', fill_value=0)
```

```python
#same output as group by function
countries_df.pivot_table(values='Population',index=['Region','SubRegion'], aggfunc='sum') #pivot_table used as a method
```
