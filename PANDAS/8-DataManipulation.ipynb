```python
#DATA MANIPULATION
import pandas as pd
countries_df = pd.read_csv(r'C:\Users\SAGAVERM\Desktop\DA dataset\top_10_countries.csv')
countries_df
```

```python
print(countries_df.shape) 
print(countries_df.columns)
print(countries_df.index)
```

```python
#rename function to change column/ index name
countries_df.rename(columns= {'Country / Dependency': 'Country'}, inplace = True)
countries_df
```

```python
#drop function to remove column or rows
#countries_df.drop(labels= 'Date', axis= 1, inplace=False)
#same can be performed in the below way
countries_df.drop(columns='Date', inplace=True)
countries_df
```

```python
countries_df.dtypes
```

```python
countries_df['Country/Region'] = countries_df['Region']+ '/' + countries_df['Country']
countries_df
```

```python
#changing column datatype
countries_df['Population'] = countries_df.Population.str.replace(',','')
countries_df['Population'] = countries_df.Population.astype(int)
```

```python
countries_df['Population (mil)'] = countries_df['Population']/1000000
countries_df
```

```python
#to remove the % sign from column
countries_df['% of world']=countries_df['% of world'].apply(lambda x: x[:-1])
countries_df
```

```python
countries_df['% of world'] = countries_df['% of world'].astype(float)
countries_df['Population'] = countries_df['Population'].astype(float)

```

```python
countries_df['Population/%of world'] = ((countries_df['Population'])/(countries_df['% of world']/100))
countries_df['Population/%of world'] = countries_df['Population/%of world'].apply(lambda x: '{:.2f}'.format(x))
countries_df
```
