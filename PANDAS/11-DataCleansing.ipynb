```python
#DATA CLEANSING
import pandas as pd
```

```python
employees = ['employee1', '  employee2', 'employee3  ', 'employee4  ','employee5', 'employee6', 'employee7', 'employee8']
position = ['analyst', 'manager', 'analyst', 'analyst', 'manager','senior manager','manager',None]
salary = [30000,56000,28000,33000,60000,75000, None, None]
columns = ['employee', 'position', 'salary']
```

```python
df1 = pd.DataFrame(data=list(zip(employees,position,salary)), columns= columns)
```

```python
df1['employee'] = df1['employee'].apply(lambda x: x.strip())
```

```python
df1['employee']
```

```python
#isna, isnull function
df1.isna() #df1.isnull(), is an alias of isna
```

```python
#dropna
df1.dropna(thresh=2) #thresh specifies the minimum number of NaN values
```

```python
df1.dropna(axis=1, thresh=7) #can also specify axis
```

```python
#fillna
df1.fillna(0) #fills 0 in all NaN values
```

```python
df1.filll() #forward fill
```

```python
df1['salary'] = df1['salary'].fillna(value = df1['salary'].mean()) #return mean salary value to NaN values
```
