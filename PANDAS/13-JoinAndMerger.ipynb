```python
import pandas as pd
emp = pd.read_excel(r'C:\Users\SAGAVERM\Desktop\DA dataset\employees_hr.xls', sheet_name='employees')
dept = pd.read_excel(r'C:\Users\SAGAVERM\Desktop\DA dataset\employees_hr.xls', sheet_name='departments')
```

```python
#join operations
emp.join(dept.set_index('id'), on='dept_id') #by default it returns left join
```

```python
#outer join and how to return sepcifc columns
emp.join(dept.set_index('id'), on = 'dept_id', how = 'outer')[['emp_id','dept_name']] #outer join
```

```python
#merge function
pd.merge(emp, dept, left_on= 'dept_id', right_on='id', how= 'left') #left join, by default it is inner join
```

```python
#to return specific columns
pd.merge(emp, dept, left_on='dept_id', right_on='id')[['emp_id','dept_name']]
```
