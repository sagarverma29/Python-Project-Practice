```python
import pandas as pd
import datetime as dt
tfl = pd.read_csv(r'C:\Users\SAGAVERM\Desktop\DA dataset\tfl-daily-cycle-hires.csv')
```

```python
tfl
```

```python
tfl['Day'] = pd.to_datetime(tfl['Day'], format='%d/%m/%Y')
```

```python
tfl.sort_values('Day', inplace=True)
```

```python
tfl['Day'] = tfl['Day'].dt.strftime('%Y-%m')
```

```python
tfl.rename({'Day': 'Month|Year'}, axis =1, inplace=True)
```

```python
tfl.drop('Unnamed: 2',axis =1,  inplace=True)
```

```python
tfl_12 = tfl.groupby('Month|Year').sum().tail(12).reset_index()
```

```python
tfl_12
```

```python
#expanding method
tfl_12['cumulative']=tfl_12['Number of Bicycle Hires'].expanding().sum()
```

```python
#rolling function
tfl_12['Rolling avg 3'] = tfl_12['Number of Bicycle Hires'].rolling(3).sum()
```

```python
tfl_12['Rolling avg 6'] = tfl_12['Number of Bicycle Hires'].rolling(6).sum()
```

```python
tfl_12['Rolling avg 3'] = tfl_12['Rolling avg 3'].fillna(value =0)
```
