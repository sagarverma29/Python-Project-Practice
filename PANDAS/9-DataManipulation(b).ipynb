```python
# deleting the unnamed column from the dataframe
tfl.drop(columns= 'Unnamed: 2', inplace= True)
```

```python
#changing the day column data type
tfl['Day'] = tfl['Day'].astype('datetime64[ns]')
#the same can be executed by using the to_datetime function
tfl['Day'] = pd.to_datetime(tfl["Day"], format='%d/%m/%Y')
```

```python
#sorting in dataframe
tfl.sort_values(by='Number of Bicycle Hires', ascending=False, inplace=True)
```

```python
#creating new column for datetime fucntion output
tfl['Month-Year']=tfl['Day'].dt.strftime('%b-%y') 
```
