```python
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.set_title('First')
ax2.set_title('second')
```

```python
fig, [[ax1,ax2],[ax3,ax4]] = plt.subplots(2,2, tight_layout = True)
ax1.set_title('ax1')
ax2.set_title('ax2')
ax3.set_title('ax3')
ax4.set_title('ax4')
```
