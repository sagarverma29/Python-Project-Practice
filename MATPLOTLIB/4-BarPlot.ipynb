```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\SAGAVERM\Desktop\DA dataset\top_10_countries.csv")
```

```python
df['Population'] = df['Population'].str.replace(',','')
```

```python
df.rename(columns={'Country / Dependency': 'Country'}, inplace=True)
```

```python
df
```

```python
df['Population'] = df['Population'].astype(int)
```

```python
df_1 = df[['Region','Population']].groupby(by=['Region']).sum()
```

```python
df_1.sort_values('Population', ascending=False, inplace=True)
```

```python
df_1.reset_index(inplace=True)
```

```python
df_1
```

```python
#barplot
fig , axes = plt.subplots(nrows=1,ncols=1,figsize=(5,5))
#axes
axes.set_title('Top Region by Population')
axes.set_xlabel('Regions')
axes.set_ylabel('Population')
#bar plot
plt.bar(x=df_1['Region'], height= df_1['Population'], color='#5584AC', edgecolor = 'black')
```

```python
#barh plot
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize= (20,8))
#axes
ax1.set_title('Total population by region')
ax1.set_xlabel('region')
ax1.set_ylabel('population')
ax2.set_title('Total population by countries')
ax2.set_xlabel('population')
ax2.set_ylabel('country')

#bar plot
ax1.bar(x = df_1['Region'], height= df_1['Population'], color= '#5584AC', edgecolor= 'black')
#barh plot
ax2.barh(y= df['Country'], width= df['Population'], color = '#5584AC', edgecolor = 'black')
```
