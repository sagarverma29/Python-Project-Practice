```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
```

```python
#adding axes to the figure 
fig = plt.figure(figsize=(5,5)).add_axes(rect=[0,0,0.5,0.5])
```

```python
#the above can also be written in this way
fig = plt.figure(figsize=(5,5))
#two axes are processed but we can see only one, because both the axes are of same dimension and are overlapping each other
axes1 = fig.add_axes(rect=[0,0,1,1])
axes2 = fig.add_axes(rect=[0,0,1,1])
```

```python
fig = plt.figure(figsize=(5,5))
axes1 = fig.add_axes(rect=[0,0,1,1])
axes2 = fig.add_axes(rect=[1.5,0,1,1])
```

```python
#axes within an axes
fig = plt.figure(figsize=(5,5))
axes1 = fig.add_axes(rect=[0,0,1,1])
axes2 = fig.add_axes(rect=[0.1,0.1,0.5,0.5])
```

```python
#set_xlim() and set_ylim()
fig = plt.figure(figsize=(5,5))
axes1 = fig.add_axes(rect=[0,0,1,1])
axes1.set_xlim([0,10]) #changes x axes limit to 0 to 10
axes1.set_ylim([0,100]) #changes y axes limit to 0 to 100
```

```python
#twinx(), adds a new y axis on the same x axis on the right hand side
fig = plt.figure(figsize=(5,5))
axes1 = fig.add_axes(rect=[0,0,1,1])
axes3 = axes1.twinx()
axes3.set_ylim([0,100]) #can change the limit of the new y-axis
```

```python
fig = plt.figure(figsize=(5,5))
axes1 = fig.add_axes(rect = [0,0,1,1])
axes1.set_title('PLOT')
axes1.set_ylabel('Y')
axes1.set_xlabel('X')
```
