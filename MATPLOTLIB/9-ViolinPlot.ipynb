```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
players = pd.read_csv(r"C:\Users\SAGAVERM\Desktop\DA dataset\players_21.csv")
```

```python
inter = players[players['club_name'] == 'Inter']['overall']
```

```python
Dortmund = players[players['club_name'] == 'Borussia Dortmund']['overall']
```

```python
data = [inter, Dortmund]
```

```python
#plot
fig = plt.figure(figsize = (5,5))
#axes
axes = fig.add_axes([0,0,1,1])
plt.xticks([1,2],['Inter','Dortmund'])
#violin plot
axes.violinplot(data)
plt.show()
```

```python
#box plot and violin plot
#plot
fig = plt.figure(figsize = (5,5))
#axes
axes = fig.add_axes([0,0,1,1])
plt.xticks([1,2],['Inter', 'Dortmund'])
#box and violin plot
axes.boxplot(data)
axes.violinplot(data)
plt.show()
```
