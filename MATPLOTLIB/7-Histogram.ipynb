```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
players = pd.read_csv(r"C:\Users\SAGAVERM\Desktop\DA dataset\players_21.csv")
```

```python
#plot
fig = plt.figure()
#axes
axes = fig.add_axes([0,0,1,1])
axes.set_ylabel('number of players')
axes.set_xlabel('overall score')
axes.hist(players['overall'], edgecolor='black')
plt.show()
```

```python
#can change the number of bins in the histogram
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.set_ylabel('number of players')
axes.set_xlabel('overall score')
#paasing scalar value into the bin
axes.hist(players['overall'], bins=20, edgecolor='black')
plt.show()
```

```python
#can also define bins by passing in a array
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.set_ylabel('number of players')
axes.set_xlabel('overall score')
#passing array like list in the bins argument
#here the labels on the x axis dont correlate with the labels.
axes.hist(players['overall'], bins= [0,25,50,75,100], edgecolor='black')
plt.show()
```

```python
#to make the labels on x correlate to the intervals in the histogram we use plt.xticks()
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.set_ylabel('number of players')
axes.set_xlabel('overall score')
#xticks method to correlate labels and intervals
plt.xticks([0,25,50,75,100], [0,25,50,75,100])
#hist plot
axes.hist(players['overall'], bins = [0,25,50,75,100], edgecolor='black')
plt.show()
```

```python
#can also change the orientation of histogram from vertical to horizontal
#plot
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.set_ylabel('number of players')
axes.set_xlabel('overall score')
#labels
plt.xticks([0,25,50,75,100], [0,25,50,75,100])
#hist
axes.hist(players['overall'], bins= [0,25,50,75,100], edgecolor= 'black', orientation= 'horizontal')
plt.show()
```
