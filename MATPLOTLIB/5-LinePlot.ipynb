```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
```

```python
tfl = pd.read_csv(r"C:\Users\SAGAVERM\Desktop\DA dataset\tfl-daily-cycle-hires.csv")
```

```python
tfl.drop(columns=['Unnamed: 2'], axis=1, inplace=True)
```

```python
tfl['Day'] = pd.to_datetime(tfl['Day'], format='%d/%m/%Y')
```

```python
tfl['Day'] = tfl['Day'].astype('datetime64[ns]')
```

```python
tfl.sort_values('Day', inplace=True)
```

```python
tfl['Year'] = tfl['Day'].dt.strftime('%Y')
```

```python
tfl_1 = tfl[['Year','Number of Bicycle Hires']]
```

```python
tfl_1 = tfl_1.groupby(by='Year').sum().reset_index()
```

```python
tfl_1
```

```python
# plot area
fig = plt.figure()
#axes
axes = fig.add_axes([0,0,1,1])
#line plot
axes.plot(tfl_1['Year'], tfl_1['Number of Bicycle Hires'])
plt.show()
```

```python
# show two different line plot
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
#double line plot
axes.plot(tfl_1['Year'], tfl_1['Number of Bicycle Hires'], tfl_1['Number of Bicycle Hires']/10)
plt.show()
```

```python
#can also change color of line plot, change its shape and inlcude markers and change opacity
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(tfl_1['Year'], tfl_1['Number of Bicycle Hires'], linestyle = '--', marker = 'o', color = '#5584AC', alpha= 0.5)
plt.show()
```
