```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.arange(0,10,0.1)
```

```python
x.shape #shape of ndarray
```

```python
y1 = x
y2 = x**2
y3 = x**3
```

```python
#plot
fig = plt.figure(figsize = (5,5))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y1)
axes.plot(x,y2)
axes.plot(x,y3)
plt.show()
```

```python
fig, [[ax1,ax2],[ax3,ax4]] = plt.subplots(2,2)
```

```python
fig, [[ax1,ax2],[ax3,ax4]] = plt.subplots(2,2)
plt.tight_layout() #can directly call as a method
```

Labels and font size

```python
fig = plt.figure(figsize = (5,5))
axes = fig.add_axes([0,0,1,1])
#plotting
axes.plot(x,y1)
axes.plot(x,y2)
axes.plot(x,y3)
#labels and fnt size
axes.set_title('Chart', loc= 'center', fontsize=20, color='blue', fontname='arial')
axes.set_ylabel('y axis', rotation='vertical')
axes.set_xlabel('x axis', rotation='vertical')
plt.xticks([0,2,4,6,8,10],['zero','two','four','six','eight','ten'], rotation='vertical')
plt.show()
```

LEGENDS

```python
fig = plt.figure(figsize=(5,5))
axes = fig.add_axes([0,0,1,1])
#labels
axes.plot(x,y1, label= 'x*y')
axes.plot(x,y2, label= 'x**y')
axes.plot(x,y3, label= 'x***y')
#the legend methods shows the label
#the position of the legend can be mentioned by passing loc argument
axes.legend(loc = 'center') #or can use scalar value from 0 to 9, or string like 'upper right'
plt.show()
```

LEGENDS WITH CATEGORICAL PLOTS

```python
df = pd.read_csv(r"C:\Users\SAGAVERM\Desktop\DA dataset\top_10_countries.csv")
```

```python
df.head(2)
```

```python
df['Population'] = df['Population'].str.replace(',','')
df['Population'] = df['Population'].astype(int)
```

```python
fig = plt.figure(figsize=(10,10))
axes = fig.add_axes([0,0,1,1])

for country in df['Country / Dependency']:
    # Fetching the population value for the specific country
    population = df[df['Country / Dependency'] == country]['Population'].values[0]
    
    # Plotting the bar for each country with its population
    axes.bar(country, population, label=country)

plt.xticks(rotation='vertical')
plt.legend()
plt.show()
```

GRID

```python
fig = plt.figure(figsize=(5,5))
axes = fig.add_axes([0,0,1,1])
#labels
axes.plot(x,y1, label= 'x*y')
axes.plot(x,y2, label= 'x**y')
axes.plot(x,y3, label= 'x***y')
axes.legend() 
#grid method
plt.grid()
plt.show()
```

```python
fig = plt.figure(figsize=(5,5))
axes = fig.add_axes([0,0,1,1])
#labels
axes.plot(x,y1, label= 'x*y')
axes.plot(x,y2, label= 'x**y')
axes.plot(x,y3, label= 'x***y')
axes.legend() 
#grid method
plt.grid(axis='x', linestyle='--', color ='grey') #can specify the axis also for the grid and the linestyle and color
plt.show()
```
