```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
players = pd.read_csv(r"C:\Users\SAGAVERM\Desktop\DA dataset\players_21.csv")
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.boxplot(players['overall'], tick_labels=['overall score'])
plt.show()
```

```python
#can remove the outliers by showfliers = False
fig = plt.figure(figsize = (5,5))
axes = fig.add_axes([0,0,1,1])
#vert argument sets the orientation of the box plot
axes.boxplot(players['overall'], tick_labels=['overall score'], showfliers= False, vert=False)
plt.show()
```

```python
inter = players[players['club_name']== 'Inter']['overall']
```

```python
dortmund = players[players['club_name']== 'Borussia Dortmund']['overall']
```

```python
data = [inter,dortmund]
```

```python
# passing data into variable
fig = plt.figure(figsize = (5,5))
axes = fig.add_axes([0,0,1,1])
#boxplot
axes.boxplot(data, tick_labels=['inter', 'dortmund'], notch=True, patch_artist=True, boxprops= dict(facecolor = 'lightgreen'), medianprops= dict(color= 'red'))
plt.show()
```
