```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
players = pd.read_csv(r"C:\Users\SAGAVERM\Desktop\DA dataset\players_21.csv")
```

```python
players.head()
```

```python
pd.set_option('display.max_columns', None)
```

```python
players_100 = players.head(100)
```

```python
players_100.head()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.set_title('overall ability vs wages')
axes.set_xlabel('overall ability')
axes.set_ylabel('wages')
#scatter plot
axes.scatter(players_100['overall'], players_100['wage_eur'])
plt.show()
```

```python
#can change size according to value of another column, which will then show the size comparasion between values
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.set_title('overall ability vs wages')
axes.set_ylabel('wages')
axes.set_xlabel('overall ability')
#scatter plot
axes.scatter(players_100['overall'], players_100['wage_eur'], s= players_100['value_eur']/200000)
#the points differ in size according to the value_eur column for each player
plt.show()
```
