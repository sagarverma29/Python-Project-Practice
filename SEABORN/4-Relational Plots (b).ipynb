```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
penguins.isna().sum()
```

```python
#scatter plot
sns.scatterplot(data= penguins, x= 'body_mass_g', y= 'flipper_length_mm', hue= 'island')
plt.show()
```

```python
sns.scatterplot(data= penguins, x= 'body_mass_g', y= 'flipper_length_mm', hue= 'sex')
plt.show()
```

```python
#using relplot to show above data
sns.relplot(data=penguins, x= 'body_mass_g', y= 'flipper_length_mm', row='sex', col= 'island', kind='scatter', hue= 'island', palette='coolwarm')
plt.show()
```
