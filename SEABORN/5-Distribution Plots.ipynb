```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

RUG PLOT

https://seaborn.pydata.org/generated/seaborn.rugplot.html#seaborn.rugplot

```python
#rugplot
fig = plt.figure(figsize=(5,5))
ax = fig.add_axes([0,0,1,1])
ax.set_ylim(165,235)
sns.rugplot(data= penguins, x= 'body_mass_g', y= 'flipper_length_mm')
plt.show()
```

```python
#rug plots are often used in combination with other plots, like below
fig = plt.figure(figsize=(5,5))
ax = fig.add_axes([0,0,1,1])
ax.set_ylim(165,235)
sns.rugplot(data= penguins, x= 'body_mass_g', y= 'flipper_length_mm', hue= 'sex')
sns.scatterplot(data= penguins, x= 'body_mass_g', y= 'flipper_length_mm', hue= 'sex')
plt.legend(loc= 'upper center')
plt.show()
```

HISTOGRAM/ HIST PLOT

https://seaborn.pydata.org/generated/seaborn.histplot.html#seaborn.histplot

```python
sns.histplot(data= penguins, x= 'body_mass_g')
plt.show()
```

```python
#with hue parameter
sns.histplot(data= penguins, x= 'body_mass_g', hue= 'sex')
plt.show()
#because of color overlapping it is not that much clear, so we use hue in y axis
```

```python
# both x and y axis
#exmaple of bivariate histogram
sns.histplot(data= penguins, x= 'body_mass_g', y= 'sex')
plt.show()
#the darker the color means more concentration 
```

```python
#similarly we can do this for islands also
#example of bivariate histogram
sns.histplot(data= penguins, x= 'body_mass_g', y= 'island')
plt.show()
```

```python
#kde parameter
sns.histplot(data= penguins, x= 'body_mass_g', kde=True)
plt.show()
```

KDE PLOT

https://seaborn.pydata.org/generated/seaborn.kdeplot.html#seaborn.kdeplot

Kernel Density Estimation by Matthew Conlen: https://mathisonian.github.io/kde/

```python
#kde plot
sns.kdeplot(data= penguins, x= 'body_mass_g')
plt.show()
```

```python
#hue parameter
sns.kdeplot(data= penguins, x= 'body_mass_g', hue= 'island')
plt.show()
```

```python
#other parameters
sns.kdeplot(x= 'body_mass_g', y= 'flipper_length_mm', data= penguins, fill=True)
plt.show()
```

```python
#kde plot with scatter plot
sns.kdeplot(data= penguins, x= 'body_mass_g', y= 'flipper_length_mm', fill= True)
sns.scatterplot(data= penguins, x= 'body_mass_g', y= 'flipper_length_mm', hue= 'island')
plt.show()
```

```python
#kde plot with rugplot
sns.kdeplot(data= penguins, x= 'body_mass_g', y= 'flipper_length_mm', fill= True)
sns.rugplot(data= penguins, x= 'body_mass_g', y= 'flipper_length_mm', hue= 'island')
plt.show()
```

ECDF PLOT (EMPIRICAL CUMULATIVE DISTRIBUTION FUCNTION)

https://seaborn.pydata.org/generated/seaborn.ecdfplot.html#seaborn.ecdfplot

```python
sns.ecdfplot(data= penguins, x= 'body_mass_g')
plt.show()
```

```python
sns.ecdfplot(data= penguins, x= 'body_mass_g', hue= 'sex')
plt.show()
```

```python
sns.ecdfplot(data= penguins, y= 'flipper_length_mm', hue= 'island')
plt.show()
```

DISPLOT

A displot can create distribution plots across a Facet Grid based on col and row variables
https://seaborn.pydata.org/generated/seaborn.displot.html

```python
sns.displot(data= penguins, x= 'body_mass_g', row='sex',col= 'island', hue = 'island')
plt.show()
```

```python
sns.displot(data= penguins, x= 'flipper_length_mm', row= 'sex', col= 'island', hue= 'sex', kind= 'kde')
plt.show()
```
