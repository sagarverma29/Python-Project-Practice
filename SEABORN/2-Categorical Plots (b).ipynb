```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
players = pd.read_csv(r"C:\Users\SAGAVERM\Desktop\DA dataset\players_21.csv")
```

```python
pd.set_option('display.max_columns', False)
```

```python
players_100 = players.head(100) #only top 100 entries
```

```python
players_100.head(2)
```

```python
#BOX PLOT
plt.figure(figsize=(10,10))
sns.boxplot(y=players_100['overall'], x=players_100['league_name'])
plt.show()
```

```python
#VIOLIN PLOT
plt.figure(figsize=(10,10))
sns.violinplot(y=players_100['overall'], x=players_100['league_name'],color='#5584AC')
plt.show()
```

```python
#STRIP PLOT
plt.figure(figsize=(10,10))
sns.stripplot(x=players_100['league_name'], y=players_100['overall'])
plt.show()
```

```python
#strip and box plot
plt.figure(figsize=(10,10))
sns.stripplot(x=players_100['league_name'], y=players_100['overall'], color='black')
sns.boxplot(x=players_100['league_name'],y=players_100['overall'], hue=players_100['league_name'], legend=True)
plt.show()
```

```python
#SWARM PLOT
plt.figure(figsize=(10,10))
sns.swarmplot(x=players_100['league_name'], y=players_100['overall'], hue= players_100['league_name'])
plt.show()
```

```python
#POINT PLOT
sns.pointplot(x=players_100['overall'])
plt.show()
```

```python
#x and y args
sns.pointplot(x=players_100['overall'], y=players_100['preferred_foot'])
plt.show()
```

```python
#point plot with hue
sns.pointplot(x=players_100['overall'], y=players_100['league_name'], hue=players_100['preferred_foot'])
plt.show()
```
