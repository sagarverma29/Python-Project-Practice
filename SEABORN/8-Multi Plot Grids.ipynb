```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

FACETGRID

https://seaborn.pydata.org/generated/seaborn.FacetGrid.html#seaborn.FacetGrid

```python
#relplot
sns.relplot(data= penguins, x= 'body_mass_g', y= 'flipper_length_mm', row= 'species', col= 'island', hue= 'sex', height =3)
plt.show()
```

```python
# You can also use FacetGrid directly to create a FacetGrid object
#... and plot charts on the grids
g= sns.FacetGrid(data= penguins, row= 'species', col= 'island')
g.map_dataframe(sns.scatterplot, x= 'body_mass_g', y= 'flipper_length_mm', hue= 'sex')
plt.show()
```

PAIRPLOT

https://seaborn.pydata.org/generated/seaborn.pairplot.html#seaborn.pairplot

```python
#plotting pair plot
sns.pairplot(data= penguins)
plt.show()
```

```python
# pairplot showing kde plots
sns.pairplot(data= penguins, kind= 'kde')
plt.show()
```

```python
#can select the diagonal plot kind seperately
sns.pairplot(data= penguins, diag_kind= 'kde', kind= 'scatter')
plt.show()
```

```python
#pairplot with hue parameter and removing corner
sns.pairplot(data= penguins, kind= 'kde', hue= 'species', corner= True)
plt.show()
```

PAIRGRID

https://seaborn.pydata.org/generated/seaborn.PairGrid.html#seaborn.PairGrid

```python
#plotting a pairgrid
g=sns.PairGrid(data= penguins)
g.map(sns.scatterplot)
plt.show()
```

```python
#can also plot different plot in upper/lower triangle and diagonal also
g=sns.PairGrid(data= penguins)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot)
plt.show()
```

```python
#passing the hue parameter
g=sns.PairGrid(data= penguins, hue= 'species')
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot)
plt.show()
```

JOINTPLOT

https://seaborn.pydata.org/generated/seaborn.jointplot.html#seaborn.jointplot

```python
#plotting joint plot
sns.jointplot(data= penguins, x= 'bill_length_mm', y= 'body_mass_g')
plt.show()
```

```python
#can change the type of plot
sns.jointplot(data= penguins, x= 'bill_length_mm', y= 'body_mass_g', kind= 'hex')
plt.show()
```

```python
#can add another plot in central and margins using
#plot_joint and plot_margins
g=sns.jointplot(data= penguins, x= 'bill_length_mm', y= 'body_mass_g', hue= 'sex')
g.plot_joint(sns.kdeplot)
g.plot_marginals(sns.rugplot, height= 0.15, color= 'red')
plt.show()
```

JOINTGRID

https://seaborn.pydata.org/generated/seaborn.JointGrid.html#seaborn.JointGrid

```python
sns.JointGrid(data= penguins, x= 'body_mass_g', y= 'flipper_length_mm').plot(sns.scatterplot, sns.kdeplot)
plt.show()
```

```python
g=sns.JointGrid(data= penguins, x= 'body_mass_g', y= 'flipper_length_mm', hue= 'sex')
g.plot_joint(sns.scatterplot)
g.plot_marginals(sns.rugplot, height=0.15)
plt.show()
```

```python
#more customization
g=sns.JointGrid()
sns.scatterplot(data= penguins, x= 'body_mass_g', y= 'flipper_length_mm', ax= g.ax_joint, hue= 'sex')
sns.kdeplot(data= penguins, x= 'body_mass_g', ax= g.ax_marg_x)
sns.histplot(data= penguins, y= 'flipper_length_mm', ax= g.ax_marg_y)
plt.show()
```
