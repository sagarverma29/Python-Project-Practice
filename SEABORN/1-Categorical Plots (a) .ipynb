```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
countries_df = pd.read_csv(r"C:\Users\SAGAVERM\Desktop\DA dataset\top_10_countries.csv")
```

```python
countries_df.rename(columns= {'Country / Dependency': 'Country'}, inplace = True)
```

```python
countries_df['Population'] = countries_df['Population'].str.replace(',','')
countries_df['Population'] = countries_df['Population'].astype(int)
```

```python
countries_df.head()
```

BAR PLOT

```python
sns.barplot(x=countries_df['Region'], y=countries_df['Population'])
```

```python
#to remove the confidence interval put, ci=none, in latest version use errorbar
sns.barplot(x=countries_df['Region'], y=countries_df['Population'], errorbar=None)
```

```python
#By default the estimator is set to mean but can specify different arthematic functions
sns.barplot(x=countries_df['Region'],y=countries_df['Population'], errorbar=None, estimator= sum)
```

```python
#hue parameter
sns.barplot(data = countries_df, x='Region', y='Population', errorbar=None, estimator=sum, hue='Region')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.barplot(data = countries_df, x='Country', y='Population', errorbar=None, estimator=sum, hue='Region')
plt.xticks(fontsize=8)
plt.show()
```

```python
#plot
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(20,10), tight_layout=True)
sns.barplot(data= countries_df, x='Region', y='Population', errorbar=None, hue='Region', dodge=False, ax=ax1, legend=True)
sns.barplot(data= countries_df, x='Country', y='Population', errorbar=None, hue='Region', ax=ax2)
plt.show()
```

COUNT PLOT

```python
#counts the number of occurence
sns.countplot(x='Region', data= countries_df, hue='Region', legend=True)
plt.show()
```
