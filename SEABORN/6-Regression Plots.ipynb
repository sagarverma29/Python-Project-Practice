```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head(2)
```

CORR() Method to find out the correlation coefficient between columns

https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.corr.html

```python
penguins_corr = penguins[['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']]
```

```python
penguins_corr.head(2)
```

```python
penguins_corr.corr()
```

REGPLOT()

https://seaborn.pydata.org/generated/seaborn.regplot.html#seaborn.regplot

```python
plt.figure(figsize=(10,10))
sns.regplot(data= penguins, x= 'flipper_length_mm', y= 'bill_length_mm')
#it shows a positive correlation
plt.show()
```

```python
#can remove the shaded region i.e confidence interval by assigning it to NULL
plt.figure(figsize= (10,10))
sns.regplot(data= penguins, x= 'flipper_length_mm', y= 'bill_depth_mm', ci= None)
#neagtive correlation
plt.show()
```

LMPLOT()

https://seaborn.pydata.org/generated/seaborn.lmplot.html#seaborn.lmplot

```python
sns.lmplot(data= penguins, x= 'flipper_length_mm', y= 'body_mass_g', height= 10, aspect= 1.5)
plt.show()
```

```python
#hue parameter
sns.lmplot(data= penguins, x= 'flipper_length_mm', y= 'body_mass_g', hue= 'island', height= 10, aspect= 1.5, ci= None)
plt.show()
```

```python
#facetgrid
sns.lmplot(data= penguins, x= 'flipper_length_mm', y= 'body_mass_g', row= 'sex', col= 'island', height= 8, aspect= 1, hue= 'island', legend= True)
plt.show()
```
