```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy 
```

```python
players = pd.read_csv(r"C:\Users\SAGAVERM\Desktop\DA dataset\players_21.csv")
```

```python
pd.set_option('display.max_columns', False)
```

```python
#removing subsitute and reserve players
players = players[-players['team_position'].isin(['RES','SUB'])]
```

```python
players.head(1)
```

HEATMAP

https://seaborn.pydata.org/generated/seaborn.heatmap.html#seaborn.heatmap

```python
players_matrix = pd.pivot_table(data= players, index= 'overall', columns= 'team_position', values= 'sofifa_id', aggfunc= 'count', fill_value= 0)
```

```python
players_matrix.sort_values('overall', ascending=False, inplace=True)
```

```python
players_matrix.head(2)
```

```python
#creating a heat map
plt.figure(figsize= (10,10))
sns.heatmap(players_matrix)
plt.show()
```

```python
#can change color map and add annotations
plt.figure(figsize= (10,10))
sns.heatmap(players_matrix, annot= True, cmap = 'coolwarm')
plt.show()
```

CLUSTERMAP

https://seaborn.pydata.org/generated/seaborn.clustermap.html#seaborn.clustermap

```python
#creating clustermap
plt.figure(figsize = (10,10))
sns.clustermap(data=players_matrix, cmap= 'coolwarm')
plt.show()
```
