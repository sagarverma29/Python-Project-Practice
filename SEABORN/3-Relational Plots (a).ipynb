```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
stocks = pd.read_excel(r"C:\Users\SAGAVERM\Desktop\DA dataset\stocks.xlsx")
```

```python
stocks['Year'] = stocks['Date'].dt.strftime('%Y')
```

```python
sns.set_theme() #to reset theme to default
```

```python
stocks.head()
```

LINE PLOT

```python
plt.figure(figsize=(20,10))
sns.lineplot(data=stocks, x= 'Date', y= 'Close', hue= 'Symbol') #can change line shape using linestyle
plt.show()
```

```python
#to print each symbol on different axes
fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize=(20,5))
sns.lineplot(data=stocks[stocks['Symbol']=='FB'], x='Date', y='Close', ax=ax1 )
sns.lineplot(data=stocks[stocks['Symbol']=='AAPL'], x='Date', y='Close', ax=ax2 )
sns.lineplot(data=stocks[stocks['Symbol']=='MSFT'], x='Date', y='Close', ax=ax3 )
plt.show()
```

```python
#other way to perform the above operation, we use relplot in seaborn
#RELPLOT
sns.relplot(data=stocks, x= 'Date', y= 'Close', kind ='line', hue ='Symbol', height=10, aspect=1.5)
plt.show()
```

```python
#subplots using relplot
#splitting acorss rows
sns.relplot(data = stocks, x= 'Date', y= 'Close', row= 'Symbol', kind = 'line', height=3, aspect=3.5)
plt.show()
```

```python
#spliting across columns
sns.relplot(data = stocks, x= 'Date', y= 'Close', col= 'Symbol', kind = 'line', height=5, aspect=1)
plt.show()
```
