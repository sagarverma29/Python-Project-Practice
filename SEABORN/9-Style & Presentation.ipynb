```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.dropna(inplace=True)
```

```python
penguins.head(2)
```

```python
x= penguins['body_mass_g']
y= penguins['flipper_length_mm']
```

TITLE AND LABELS

```python
#using matplotlib
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.set_title('Body mass/ Flipper length', fontsize= 25, color='blue')
axes.set_xlabel('Body mass g', fontsize = 15)
axes.set_ylabel('Flipper length mm', fontsize= 15)
sns.scatterplot(data= penguins, x=x, y=y)
plt.show()
```

```python
#using seaborn
ax1 = sns.scatterplot(data= penguins, x=x, y=y)
ax1.set(title= 'Title', xlabel= 'Body mass g', ylabel= 'Flipper length mm')
sns.set_theme(font_scale= 1.5)
plt.show()
```

FIGURE SIZE

```python
fig = plt.figure(figsize= (10,10))
axes = fig.add_axes([0,0,1,1])
sns.scatterplot(data= penguins, x=x, y=y)
plt.show()
```

```python
#using set_context
fig= plt.figure(figsize=(6,6))
axes = fig.add_axes([0,0,1,1])
sns.set_context('poster')
sns.scatterplot(data= penguins, x=x, y=y)
plt.show()
```

```python
fig = plt.figure(figsize=(6,6))
axes= fig.add_axes([0,0,1,1])
sns.set_context('notebook')
sns.scatterplot(data= penguins, x=x, y=y)
plt.show()
```

Setting the Theme and Style

palettes https://seaborn.pydata.org/tutorial/color_palettes.html

can use various themes like, 'whitegrid', 'darkgrid', 'white', 'dark', ticks' can also specify the palette you want to use.

```python
#setting style
sns.set_style('whitegrid')
fig= plt.figure(figsize= (6,6))
axes = fig.add_axes([0,0,1,1])
axes.set_xlabel('Body mass',fontsize=15)
axes.set_ylabel('Flipper length',fontsize=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
sns.scatterplot(data= penguins, x=x, y=y)
plt.show()
```

```python
#setting style and palette
sns.set_theme(style= 'ticks', palette='colorblind')
fig = plt.figure(figsize= (6,6))
axes = fig.add_axes([0,0,1,1])
sns.scatterplot(data= penguins, x=x, y=y)
plt.show()
```

DESPINE

https://seaborn.pydata.org/generated/seaborn.despine.html

```python
sns.scatterplot(data= penguins, x=x, y=y)
sns.despine()
plt.show()
```

```python
#can also remove the left and bottom spine i.e a spineless graph
sns.scatterplot(data= penguins, x=x, y=y)
sns.despine(left=True, bottom=True)
plt.show()
```

rcParams

https://matplotlib.org/stable/api/matplotlib_configuration_api.html

```python
#rcParams, rc stands for run commands, it allows to customize matplotlib and seaborn
plt.rcParams
```

```python
sns.set_theme(style= 'ticks', rc= {'figure.figsize': (10,10), 'axes.titlesize': 20, 'axes.labelsize': 10, 'legend.loc':'center' })
```

```python
sns.scatterplot(data= penguins, x=x, y=y, hue= 'island')
plt.show()
```
